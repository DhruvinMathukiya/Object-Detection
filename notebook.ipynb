# Training Model

```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

model.train(
    data="data.yaml",
    epochs=15,
    imgsz=640,
    batch=16,
    workers=4,
    device=0,
    amp=True,
    val=False
)
```
